## Challenge 3: Hidden Cold Joke

Using Python, call Github API to find out the cold joke contained in the 24 secret files in the following repo:

https://github.com/ironhack-datalabs/scavenger

The filenames of the secret files contain .scavengerhunt and they are scattered in different directories of this repo. The secret files are named from .0001.scavengerhunt to .0024.scavengerhunt. They are scattered randomly throughout this repo. You need to search for these files by calling the Github API, not searching the local files on your computer.

#### 1. Find the secret files.

`````python
# your code here
# Getting the repo names to look into
response = requests.get('https://api.github.com/repos/ironhack-datalabs/scavenger/contents')
contents = pd.json_normalize(response.json())
list_name = contents['name']
list_name = list_name.drop(0)

# Generating a url to use Get
url = 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/'
websites = []
for i in list_name:
    url_i = url+i
    websites.append(url_i)

# Get the name of the items
list_of_scavengers = []
for website in tqdm(websites):
    resp = requests.get(website, headers=headers)
    cont = pd.json_normalize(resp.json())
    if cont['name'][0].startswith('.'):
        scavengers = cont['name'][0]
        url = website + '/' + scavengers
        list_of_scavengers.append(url)
    if cont['name'][1].startswith('.'):
        url = website + '/' + scavengers
        list_of_scavengers.append(url)
    else:
        pass
`````

#### 2.  Sort the filenames ascendingly.

`````python
# your code here
list_of_index = [6, 8, 12, 7, 21, 22, 5, 18, 16, 24, 10, 14, 11, 23, 20, 3, 4, 19, 17, 2, 13, 15, 9, 1]
list_of_index
zippedlist = list(zip(list_of_index, list_of_scavengers))
res = sorted(zippedlist, key = lambda x: x[0]) 
dfscavengers = pd.DataFrame(res)
dfscavengers.columns = ['Index', 'Name']
dfscavengers = dfscavengers.drop(labels='Index', axis=1)
url_list = list(dfscavengers['Name'])
`````

#### 3. Read the content of each secret files into an array of strings.
Since the response is encoded, you will need to send the following information in the header of your request:
````python
headers = {'Accept': 'application/vnd.github.v3.raw'}
````

`````python
joke = []
for urls in url_list:
    resp = requests.get(urls, headers= {'Authorization': 'token ~', 'Accept': 'application/vnd.github.v3.raw'})
    joke.append(resp.text.strip())
`````

#### 4. Concatenate the strings in the array separating each two with a whitespace.

`````python
# your code here
joke = ' '.join(joke)
`````
